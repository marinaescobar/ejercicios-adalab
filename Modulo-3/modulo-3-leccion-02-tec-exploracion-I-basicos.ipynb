1- Abre el fichero de `rating-and-performance` con el método que sea más correcto. Además muestra las 5 primeras filas del *DataFrame*. 

```python
rating_performance = pd.read_csv('Materiales/rating-and-performance.csv', index_col=0)

rating_performance = pd.DataFrame(rating_performance)

rating_performance.head(5)
```

2- Pongamos en práctica el `loc` e `iloc`:

  - Usando el `loc`: 

      - Extraed todas las filas de la columna "price". ¿De qué tipo es el resultado obtenido?

      - Extraed todas las filas de las columnas "price" y "retail_price". ¿De qué tipo es el resultado obtenido?

      - Extraed las filas pares, de las columnas "rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count". 

    - Extraed las las filas cuyo precio sea mayor que 10 y sus unidades vendidas ("units_sold") sea mayor que 5000. 

  - Usando el `iloc` (sacaremos los mismos resultados que con el loc): 

    - Extraed todas las filas de la columna "price". 

    - Extraed todas las filas de las columnas "price" y "retail_price". 

    - Extraed las filas pares, de las columnas "rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count". 

```python
columnas_price = rating_performance.loc[: , "price"]
print(f"Los datos de todas las filas de la columna 'price' son:\n{columnas_price}")
print(f"El tipo de datos obtenido es: {columnas_price.dtype}")

print("\n-------------------------------\n")

columnas_price_retail = rating_performance.loc[: , ["price" , "retail_price"]]
print(f"Los datos de todas las filas de las columnas 'price' y 'retail_price' son:\n{columnas_price_retail}")
print(f"El tipo de datos obtenido es: {columnas_price_retail.dtypes}")

print("\n-------------------------------\n")

filas_pares = rating_performance.loc[::2 , "rating":"rating_one_count"]
print(f"Los datos de todas las filas pares de las columnas 'rating_count' a la 'rating_one_count' son:\n{filas_pares}")

print("\n-------------------------------\n")

precios_unidades = rating_performance.loc[(rating_performance.price > 10) & (rating_performance.units_sold > 5000)]
print(f"Los datos extraídos basados en la condición de que price > 50 y units_sold > 5000 son:\n{precios_unidades}")
```

```python
columnas_price2 = rating_performance.iloc[: , 2]
print(f"Los datos de todas las filas de la columna 'price' son:\n{columnas_price2}")

print("\n-------------------------------\n")

columnas_price_retail2 = rating_performance.iloc[: , [2 , 3]]
print(f"Los datos de todas las filas de las columnas 'price' y 'retail_price' son:\n{columnas_price_retail2}")

print("\n-------------------------------\n")

filas_pares2 = rating_performance.iloc[::2 , 7:14]
print(f"Los datos de todas las filas pares de las columnas 'rating_count' a la 'rating_one_count' son:\n{filas_pares2}")
```
