6- Seleccionad solo las columnas de tipo categórico del DataFrame. Después, mostrad los valores únicos y sus frecuencias para cada una de las columnas seleccionadas. 

```python
rating_df_cat = rating_df.select_dtypes(include = "object")

rating_df_cat.drop(['crawl_month'], axis = 1, inplace=True)

rating_cat = rating_df_cat.columns

for columna in rating_cat:
    
    print(f"\n------ COLUMNA {columna.upper()} ------")
    print(f"Valores únicos:")
    print(f"{rating_df_cat[columna].unique()}\n")
    print(f"La frecuencia de estos valores es:\n{rating_df_cat[columna].value_counts()}")
```
