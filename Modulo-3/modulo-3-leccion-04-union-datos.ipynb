```python
import pandas as pd
import numpy as np
import regex as re

pd.set_option('display.max_columns', None) 
```

1. Unid los dos DataFrames (`rating-and-performance` & `insights.csv`) en uno solo, usando el método que considereis más correcto. Querremos quedarnos con todos los productos que estén en los dos conjuntos de datos. 

```python
# Importa los dataframes y les asigna dos variables

df_rating = pd.read_csv("Materiales/rating-and-performance.csv" , index_col=0)

df_insights = pd.read_csv("Materiales/insight.csv", index_col=0)
```

```python
# En ambos casos la columna que tienen en común es 'merchant_id / merchant.id'
# Para quedarnos con los productos que están presentes en ambos conjuntos se usará merge como método de unión

df_merged = df_rating.merge(df_insights , left_on = 'merchant_id' , right_on = 'merchant.id')

df_merged.head(3)
```

2. Explorad el nombre de las columnas, ¿hay alguna que este repetida o que no nos aporte información de valor? ¿Hay alguna columna que solo tenga un valor? En caso de ser así, eliminadlas. **NOTA** Hay tres columnas que cuando exploramos sus valores únicos y sus frecuencias ayer, nos deberíamos haber dado cuenta que solo tiene un valor.  

```python
print(f"Los nombres de las columnas son:")
df_merged.columns
```

```python
# Revisa qué columnas tienen valores únicos y las elimina

for columna in df_merged.columns:
    
    if len(df_merged[columna].unique()) < 2:
        print(f"La columna {columna.upper()} tiene un valor único ({df_merged[columna].unique()[0]}) y por tanto se elimina del DataFrame")
        df_merged.drop(columna , axis = 1 , inplace = True)
```

```python
# Muestra la primera fila del DataFrame para encontrar columnas redundantes:

df_merged.head(1)
```

```python
# En base a lo observado, se crea una lista de aquellas columnas redundantes o cuyos valores se consideran de escasa relevancia de cara a un futuro análisis:
columnas_irrelevantes = ['title_orig', 'merchant.id' , 'product_url' , 'product_picture' , 'merchant_title' , 'merchant_profile_picture']

# Elimina las columnas del listado
df_merged.drop(columnas_irrelevantes, axis = 1 , inplace = True)
```

3. Algunas de las columnas tienen `.` en sus nombres y en otras tenemos `_`.  Unifica el nombre de las columnas para que todas tengan "_". Además, todas deben estar en minúsculas. 

```python
# Crea un diccionario de los nombres de las columnas y sus nuevos valores
nuevos_nombres = {columna: columna.lower().replace(".", "_") for columna in df_merged.columns}

# Reemplaza los nombres de las columnas por los datos nuevos
df_merged.rename(columns = nuevos_nombres, inplace = True)
```

4. En la columna de `merchant_info_subtitle`, si nos fijamos hay algunas celdas donde tenemos entre paréntesis el número total de calificaciones para el comerciante (que es la misma información que la que tenemos en la columna "merchant_rating_count"). Eliminad de la columna "merchant_info_subtitle" la parte del *string* de los paréntesis. Guardar el resultado de vuestra limpieza en una columna que se llame igual, es decir, "merchant_info_subtitle". 


```python
patron_regex = "\((.*?)\)"

df_merged["merchant_info_subtitle"] = df_merged["merchant_info_subtitle"].str.replace(patron_regex , '')
```

5. Si nos fijamos, en la primera fila antes solo teníamos un paréntesis, pero los hemos eliminado en el paso anterior, y ahora tenemos un *string* vacío, lo cual no tiene mucho sentido. En este ejercicio debereis reemplazar esos *strings* vacios por nulos de Pandas (**NOTA** estos deben ser indicados de la siguiente forma `np.nan`, y deberéis importar numpy). 

```python
df_merged["merchant_info_subtitle"].replace('' , np.nan , inplace = True)
```

    - Seleccionad aquellos productos cuyo color es verde y rojo. ¿Cuántos productos tenemos de cada tipo de color?

    - Seleccionad aquellos productos de la talla M, que valgan más de 15. ¿Cuántos productos tenemos?

    - Seleccionad aquellos productos cuyo precio este entre 1 y 2 (ambos inclusive). ¿Cuántos productos tenemos?

```python
# PRODUCTOS DE COLOR VERDE

# Selecciona y asigna a una variable los productos de color verde
productos_verdes = df_merged.loc[df_merged['product_color'] == 'green' , :]

print(f"La cantidad de productos de color verde es: {len(productos_verdes)}")

# Otra opción de hacerlo (consultada en internet):
print(f"La cantidad de productos de color verde (con la segunda opción) es: {df_merged['product_color'].value_counts().get('green' , 0)}")
```

```python
# PRODUCTOS DE COLOR VERDE Y ROJO

# Selecciona y asigna a una variable los productos de color verde y de color rojo
productos_verde_rojo = df_merged.loc[(df_merged['product_color'] == 'green') | (df_merged['product_color'] == 'red'), :]

print(f"La cantidad de productos de cada color es:\n{df_merged['product_color'].value_counts().get(['green' , 'red'] , 0)}")
```

```python
# PRODUCTOS TALLA M CON VALOR > 15

# Selecciona y asigna a una variable los productos de talla M con valor superior a 15
prod_M_15 = df_merged.loc[(df_merged['product_variation_size_id'] == 'M') & (df_merged['price'] > 15), :]

# Cuenta las filas
print(f"La cantidad de productos de talla M con un valor superior a 15 es: {prod_M_15.shape[0]}")
```

```python
# PRODUCTOS PRECIO ENTRE 1 Y 2

prod_1_2 = df_merged[df_merged['price'].between(1 , 2 , inclusive = 'both')]

# Cuenta las filas
print(f"La cantidad de productos cuyo precio oscila entre 1 y 2 (ambos incluidos) es: {prod_M_15.shape[0]}")
```

```python
# PRODUCTOS 1 O MÁS DÍGITOS EN SU TÍTULO

patron_regex = "\d+"

prod_digitos = df_merged[df_merged["title"].str.contains(patron_regex, regex = True, na = False)]

# Cuenta las filas
print(f"La cantidad de productos cuyo título incluye algún dígito es: {prod_digitos.shape[0]}")
```
